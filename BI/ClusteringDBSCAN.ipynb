```python
import numpy as np 
import pandas as pd
import datetime as dt
from sklearn.cluster import KMeans 
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
data = pd.read_excel('Nutrition.xlsx')
data
```

```python
f_data = data.drop(columns=['Category','Meal_Type','Calories','Food_Item'])
f_data
```

```python
p_lables = range(1,6)
p_groups = pd.qcut(data['Protein'], q=5, labels=p_lables)
ca_lables = range(1,6)
ca_groups = pd.qcut(data['Carbohydrates'], q=5, labels=ca_lables)
fa_lables = range(1,6)
fa_groups = pd.qcut(data['Fat'], q=5, labels=fa_lables)
fi_lables = range(1,6)
fi_groups = pd.qcut(data['Fiber'], q=5, labels=fi_lables)
su_lables = range(1,6)
su_groups = pd.qcut(data['Sugars'], q=5, labels=su_lables)
so_lables = range(1,6)
so_groups = pd.qcut(data['Sodium'], q=5, labels=so_lables)
c_lables = range(1,6)
c_groups = pd.qcut(data['Cholesterol'], q=5, labels=c_lables)
w_lables = range(1,6)
w_groups = pd.qcut(data['Water_Intake'], q=5, labels=w_lables)
t_lables = range(1,6)
t_groups = pd.qcut(data['Total_Calories_Per_User'], q=5, labels=t_lables)
```

```python
f_data['P'] = p_groups
f_data['CA'] = ca_groups
f_data['FA'] = fa_groups
f_data['FE'] = fi_groups
f_data['SU'] = su_groups
f_data['SO'] = so_groups
f_data['C'] = c_groups
f_data['W'] = w_groups
f_data['T'] = t_groups
```

```python
f_data
```

```python
f_data['P'] = f_data['P'].astype(int)
f_data['CA'] = f_data['CA'].astype(int)
f_data['FA'] = f_data['FA'].astype(int)
f_data['FE'] = f_data['FE'].astype(int)
f_data['SU'] = f_data['SU'].astype(int)
f_data['SO'] = f_data['SO'].astype(int)
f_data['C'] = f_data['C'].astype(int)
f_data['W'] = f_data['W'].astype(int)
f_data['T'] = f_data['T'].astype(int)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(f_data)
```

```python
for eps in np.arange(1, 3.0, 0.1):
    for min_samples in np.arange(5, 20, 1):
        for leaf_size in np.arange(10, 20, 1):
            dbscan = DBSCAN(eps=eps, min_samples=min_samples, leaf_size=leaf_size)
            dbscan.fit(X_scaled)
            labels = dbscan.labels_
            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
            n_noise = list(labels).count(-1)

            if 5 <= n_clusters <= 20:
                score = silhouette_score(X_scaled, labels)
                print(f"eps={eps:.1f}, min_samples={min_samples}, leaf_size={leaf_size} â†’ clusters={n_clusters}, noise points={n_noise}, score={score:.4f}")
```
